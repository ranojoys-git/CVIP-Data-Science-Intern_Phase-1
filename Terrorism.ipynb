```python
new_terror=terror.dropna(axis=1, thresh=120000)
new_terror
```

```python
new_terror.head()
```

```python
new_terror=new_terror.fillna("unknown")
new_terror.head()
```

```python
new_terror.columns
```

```python
new_terror = new_terror.copy()
new_terror.rename(columns={"iyear":"Year", "imonth":"Month", "iday":"Day", "country":"CountryCode", 
                                  "country_txt":"Country", "region":"RegionCode", "region_txt":"Region", 
                                  "provstate":"State", "city":"City", "latitude":"Latitude", "longitude":"Longitude",
                                  "specificity":"Specificity", "success":"Success", "attacktype1":"AttackTypeCode", 
                                  "attacktype1_txt":"AttackType", "targtype1":"TargetTypeCode", "targtype1_txt":"TargetType", 
                                  "target1":"Target", "natlty1":"Nationality", "gname":"GroupName", "weaptype1":"WeaponCode", 
                                  "weaptype1_txt":"WeaponType", "nkill":"Killed", "nwound":"Wounded"}, inplace = True)
```

```python
new_terror = new_terror[['Year', 'Month', 'Day', 'CountryCode', 'Country', 'RegionCode', 'Region', 
                                       'State', 'City', 'Latitude', 'Longitude', 'Specificity', 'Success', 
                                       'AttackTypeCode', 'AttackType', 'TargetTypeCode', 'TargetType', 'Target', 
                                       'Nationality', 'GroupName', 'WeaponCode', 'WeaponType', 'Killed', 'Wounded']]
new_terror.head()
```

```python
new_terror.columns.to_list()
```

```python
new_terror['Year'].value_counts().sort_index()
```

```python
print("Year with the most attacks: ", new_terror['Year'].value_counts().idxmax())
print("Country with most attacks: ", new_terror['Country'].value_counts().idxmax())
print("Group with the most attacks:", new_terror['GroupName'].value_counts().index[1])
```

```python
x_year = new_terror['Year'].unique()
y_year = new_terror['Year'].value_counts().sort_index().values
plt.figure(figsize=(15,10))
plt.title("Number of Attacks per Year")
plt.xlabel("Attack Years",fontsize=15)
plt.ylabel("Number of Attacks each year")
plt.xticks(rotation=45)
sns.barplot(x=x_year, y=y_year, palette= 'rocket') 
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sns.countplot('Year', data=new_terror, palette='RdYlGn_r',edgecolor=sns.color_palette("YlOrBr", 5))
plt.ylabel("Count of Terrorist Activities")
plt.xticks(rotation=45)
plt.title('CountPLot of Number Of Terrorist Activities Each Year')
plt.show()
```

```python
pd.crosstab(new_terror.Year,new_terror.Region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist activites by Region in each year')
plt.ylabel("Number of attacks")
plt.show()
```

```python
new_terror['Wounded'] = new_terror['Wounded'].replace('unknown', 0)
new_terror['Wounded'] = new_terror['Wounded'].fillna(0).astype(int)
new_terror['Killed'] = new_terror['Killed'].replace('unknown', 0)
new_terror['Killed'] = new_terror['Killed'].fillna(0).astype(int)
new_terror['casualities'] = new_terror['Killed'] + new_terror['Wounded']
```

```python
terror_copy = new_terror.sort_values(by='casualities',ascending=False)[:30]
```

```python
new_terror.corr()
```

```python
plt.figure(figsize=[15,8])
sns.heatmap(new_terror.corr(),cmap='RdYlGn', linewidths=0.4)
plt.show()
```

```python
attack_country = new_terror.Country.value_counts()[:15]
attack_country
```

```python
plt.subplots(figsize=(12,6))
sns.barplot(attack_country.index,attack_country.values,palette='Blues_d')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation= 45)
plt.show()
```

```python
attack_groups = new_terror.GroupName.value_counts()[1:15]
attack_groups
```

```python
plt.subplots(figsize=(12,6))
sns.barplot(attack_groups.index,attack_groups.values,palette=sns.color_palette("hls", 8))
plt.title('Groups involving most in terrorism')
plt.xlabel('Groups')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

```python
group_killed= new_terror[['GroupName','Killed']].groupby(['GroupName'],axis=0).sum().sort_values('Killed', ascending=False).head(20)
group_killed
```

```python
plt.subplots(figsize=(12,6))
sns.barplot(group_killed.index, group_killed.Killed.values,palette="crest")
plt.title('Number of People Killed by each terrorist group')
plt.xlabel('Terrorist Groups')
plt.ylabel('Number of people killed')
plt.xticks(rotation= 90)
plt.show()
```

```python
attack_type = new_terror.AttackType.value_counts()[:15]
attack_type
```

```python
plt.subplots(figsize=(12,6))
sns.barplot(x=attack_type.index,y=attack_type.values,palette="inferno")
plt.title('Attack Types')
plt.xlabel('Types ot attacks')
plt.ylabel('Number of attacks')
plt.xticks(rotation= 90)
plt.show()
```

```python
attack_wounded = new_terror[new_terror['Wounded'] != 'Unknown'][['AttackType','Wounded']].groupby(["AttackType"],axis=0).sum()
attack_wounded
```

```python
plt.subplots(figsize=(15,6))
sns.barplot(x=attack_wounded.index, y=attack_wounded.Wounded.values,palette="viridis")
plt.title('Number of People Wounded in each Attack Type', fontsize=15)
plt.xlabel('Atack Types', fontsize=12)
plt.ylabel('Number of People Wounded', fontsize=12)
plt.xticks(rotation= 90)
plt.show()
```
